```python
# Sample data representing the average temperature (in degrees Celsius) from 2010 to 2014
data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Avg_Temperature': [14.1, 14.3, 14.5, 14.2, 14.6]}
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame(data)
df
# VERSION ONE

plt.plot(df['Year'], df['Avg_Temperature'], marker='o') # first value is X, second is Y

plt.title('Avarage temperature from 2010 to 2014') # we add title
plt.xlabel('Year') # title on x axis
plt.ylabel('Temperatue') # title on y axis

plt.xticks(df['Year']) #year values are rounded
plt.grid(True) # grid(ruudustik)
plt.show() # show the visualization
```

```python
# VERSION TWO, create a plot

df.plot(x='Year', y='Avg_Temperature', kind='line')
plt.title('Average temperature from 2010 to 2014') #we added title
plt.xlabel('Year') #title on x axis
plt.ylabel('Temperature') #title on y axis

plt.xticks(df['Year']) #year values are rounded
plt.grid(True)
plt.show() #show you the visualization
```

```python
# BAR CHART

plt.bar(df['Year'], df['Avg_Temperature'], color='skyblue')
plt.title('Avg temperature from 2010 to 2014')
plt.xlabel('Year')
plt.ylabel('Temperature')
```

```python
# HORIZONTAL BAR CHART

plt.barh(df['Year'], df['Avg_Temperature'], color='orange')
plt.title('Avg temperature from 2010 to 2014')
plt.xlabel('Temperature')
plt.ylabel('Year')
```

```python
# HISTOGRAM - presents the distribution of a single variable, frequency, represents numerical data

data2 = {'Scores': [85, 92, 78, 20, 89, 90, 82, 56, 60, 23, 91, 40, 77]}
df2 = pd.DataFrame(data2)
df2

plt.hist(df2['Scores'], bins=5, color='green', edgecolor='black')
```

```python
# SCATTER PLOT (hajuvusdiagramm)- used to visualize the relationship between two variables

data3 = {'X': [1,2,3,5,5,2], 'Y':[5,8,7,4,9,5]}

df3 = pd.DataFrame(data3)
df3

plt.scatter(x = df3['X'], y=df3['Y'])
plt.title('Scatter Plot')
plt.grid(True) # annab ruudustiku
```

```python
# PIE CHART - represents parts of a whole

data4 = {'Category': ['A', 'B', 'C', 'D'], 'Values': [15, 30, 25, 40]}
df4 = pd.DataFrame(data4)
df4

plt.pie(df4['Values'], labels = df4['Category'], autopct='%1.1f%%')
plt.show()
```

```python
# BOX PLOT - distribution of a dataset and identifies outliers

plt.boxplot(df4['Values'])
plt.grid(True)
plt.show()
```
